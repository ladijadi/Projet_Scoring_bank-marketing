# Projet: Bank Marketing Scoring

## Contexte du projet

Les institutions financières mènent régulièrement des campagnes de **marketing direct** afin de promouvoir leurs produits financiers auprès de leur clientèle.  
Ces campagnes, souvent réalisées par téléphone, engendrent des coûts opérationnels importants et nécessitent un **ciblage efficace** afin de maximiser leur rentabilité.

Dans ce contexte, la mise en place de modèles prédictifs permet d’identifier les clients les plus susceptibles de souscrire à une offre donnée, et ainsi d’optimiser la prise de décision marketing.

Ce projet s’appuie sur le jeu de données **Bank Marketing**, issu du **UCI Machine Learning Repository**, qui regroupe des informations relatives à des campagnes de marketing direct menées par une institution bancaire portugaise.

---

## Objectif du projet

L’objectif de ce projet est de construire un **modèle de scoring d’appétence client** permettant de prédire la probabilité qu’un client souscrive à un **dépôt à terme** à l’issue d’une campagne marketing bancaire.

Plus précisément, je cherche à :
- analyser les caractéristiques des clients contactés lors des campagnes,
- identifier les variables influençant la décision de souscription,
- comparer plusieurs modèles de **classification supervisée**,
- sélectionner et analyser un modèle final exploitable dans un cadre de décision marketing.

---

## Formulation du problème de Machine Learning

La variable cible \( Y \) est binaire et indique si le client a souscrit à un dépôt à terme (`yes`) ou non (`no`).  
Le problème est donc formulé comme un **problème de classification supervisée binaire**.

D’un point de vue probabiliste, la variable \( Y \) peut être modélisée comme une **variable de Bernoulli conditionnellement aux variables explicatives \( X \)**.  
L’objectif est d’estimer la probabilité conditionnelle :

\[
P(Y = 1 \mid X = x)
\]

afin de construire un score d’appétence permettant de classer les clients selon leur probabilité de souscription.

## Présentation du jeu de données

Le jeu de données utilisé est **`bank-additional-full.csv`**, version enrichie du dataset Bank Marketing, décrite dans :

> Moro, S., Cortez, P., & Rita, P. (2014). *A Data-Driven Approach to Predict the Success of Bank Telemarketing*.  
> *Decision Support Systems*.

Il contient **41 188 observations** et **20 variables explicatives**, auxquelles s’ajoute la variable cible.

Certaines variables catégorielles comportent la modalité `"unknown"`, correspondant à des informations non renseignées.  
Ces modalités sont conservées et traitées comme une catégorie à part entière, car elles peuvent contenir une information pertinente.

La variable `duration`, correspondant à la durée du dernier appel, est **exclue du modèle**, car elle n’est connue qu’après le contact client et introduirait une fuite d’information dans un cadre prédictif réaliste.

---

## Description des variables

Les variables du dataset peuvent être regroupées en plusieurs catégories :

### Variables socio-démographiques et financières
- `age` : âge du client (numérique)
- `job` : type de profession (catégorielle)
- `marital` : statut matrimonial (catégorielle)
- `education` : niveau d’éducation (catégorielle)
- `default` : existence d’un défaut de crédit (catégorielle)
- `housing` : prêt immobilier en cours (catégorielle)
- `loan` : prêt personnel en cours (catégorielle)

### Variables liées à la campagne en cours
- `contact` : type de contact utilisé
- `month` : mois du dernier contact
- `day_of_week` : jour de la semaine du dernier contact
- `campaign` : nombre de contacts effectués durant la campagne

### Variables liées aux campagnes précédentes
- `pdays` : nombre de jours écoulés depuis le dernier contact
- `previous` : nombre de contacts avant la campagne actuelle
- `poutcome` : résultat de la campagne précédente

### Variables de contexte socio-économique
- `emp.var.rate` : taux de variation de l’emploi
- `cons.price.idx` : indice des prix à la consommation
- `cons.conf.idx` : indice de confiance des consommateurs
- `euribor3m` : taux Euribor à 3 mois
- `nr.employed` : nombre de personnes employées

### Variable cible
- **`y`** : souscription à un dépôt à terme  
  - `1` : le client a souscrit  
  - `0` : le client n’a pas souscrit  

---

## Enjeu du modèle

Un tel modèle de scoring permettrait :
- d’améliorer le **ciblage des campagnes marketing**,
- de réduire le nombre de contacts non pertinents,
- et d’optimiser la prise de décision dans un contexte bancaire.

Dans la suite du projet, je mets en place un pipeline de préparation des données, puis je compare plusieurs modèles de classification afin de sélectionner le modèle le plus pertinent.
