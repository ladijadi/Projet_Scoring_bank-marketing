# Définition du problème - Bank Marketing

## 1. Contexte général

Le dataset **Bank Marketing** provient de campagnes de marketing direct menées par une institution bancaire portugaise entre 2008 et 2010.  
L’objectif de ces campagnes était de proposer un **dépôt à terme** à des clients préalablement contactés.

Ce jeu de données est largement utilisé dans la littérature en data mining et a été décrit notamment dans :

> Moro, S., Cortez, P., & Rita, P. (2014).  
> *A Data-Driven Approach to Predict the Success of Bank Telemarketing*.  
> Decision Support Systems.

---

## 2. Objectif du projet

L’objectif est de construire un **modèle de classification supervisée** permettant d’estimer la probabilité qu’un client souscrive à un dépôt à terme.

Il s’agit donc de prédire une variable binaire à partir d’un ensemble de variables explicatives hétérogènes (numériques et catégorielles).

---

## 3. Formulation probabiliste

La variable cible \( Y \) prend les valeurs :

\[
Y =
\begin{cases}
1 & \text{si le client souscrit} \\
0 & \text{sinon}
\end{cases}
\]

Le problème consiste à estimer la probabilité conditionnelle :

\[
P(Y = 1 \mid X = x)
\]

où \( X \in \mathbb{R}^p \) représente le vecteur des variables explicatives.

On suppose que, conditionnellement à \( X \), la variable \( Y \) suit une loi de Bernoulli de paramètre \( p(x) \).

L’objectif est donc d’estimer une fonction :

\[
f : \mathbb{R}^p \rightarrow [0,1]
\]

telle que :

\[f(x) \approx P(Y = 1 \mid X = x)\]

---

## 4. Description du jeu de données

Le fichier utilisé est :

**`bank-additional-full.csv`**

Il contient :

- **41 188 observations**
- **20 variables explicatives**
- **1 variable cible**

Les variables du dataset peuvent être regroupées en plusieurs catégories :

### Variables socio-démographiques et financières
- `age` : âge du client (numérique)
- `job` : type de profession (catégorielle)
- `marital` : statut matrimonial (catégorielle)
- `education` : niveau d’éducation (catégorielle)
- `default` : existence d’un défaut de crédit (catégorielle)
- `housing` : prêt immobilier en cours (catégorielle)
- `loan` : prêt personnel en cours (catégorielle)

### Variables liées à la campagne en cours
- `contact` : type de contact utilisé
- `month` : mois du dernier contact
- `day_of_week` : jour de la semaine du dernier contact
- `campaign` : nombre de contacts effectués durant la campagne

### Variables liées aux campagnes précédentes
- `pdays` : nombre de jours écoulés depuis le dernier contact
- `previous` : nombre de contacts avant la campagne actuelle
- `poutcome` : résultat de la campagne précédente

### Variables de contexte socio-économique
- `emp.var.rate` : taux de variation de l’emploi
- `cons.price.idx` : indice des prix à la consommation
- `cons.conf.idx` : indice de confiance des consommateurs
- `euribor3m` : taux Euribor à 3 mois
- `nr.employed` : nombre de personnes employées

### Variable cible
- **`y`** : souscription à un dépôt à terme  
  - `1` : le client a souscrit  
  - `0` : le client n’a pas souscrit  

Certaines variables catégorielles comportent la modalité `"unknown"`, conservée comme catégorie distincte.

---

## 5. Traitement particulier de la variable `duration`

La variable `duration` correspond à la durée du dernier appel téléphonique.

Or cette information n’est disponible qu’après la fin du contact client.  
Son inclusion dans le modèle introduirait une **fuite d’information (data leakage)**.

Elle est donc exclue des modèles afin de conserver un cadre prédictif cohérent.

---

## 6. Nature du problème

Il s’agit :

- d’un **problème de classification binaire**,
- d’un dataset **déséquilibré** (~11 % de classe positive),
- d’un contexte nécessitant l’utilisation de métriques adaptées telles que **ROC-AUC**, **F1-score** et **Recall**.
